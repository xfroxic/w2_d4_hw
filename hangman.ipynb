```python
# Frank Goshko 11/4/21
# Guess the letter assignment

import random
from IPython.display import clear_output as co

print("Let's play Hangman!")

wordDict = {"Animals": ["alligator", "buffalo", "chimpanzee", "dolphin", "elephant", "flamingo", "gorilla", "horse", "iguana", "jaguar"], \
            "Candy": ["twix", "heath", "butterfinger", "milkyway", "snickers", "sweedishfish", "candycorn", "skittles", "airheads", "rolo"], \
            "Food": ["pizza", "spaghetti", "lasagne", "friedrice", "taco", "cheeseburger", "steak", "guacamole", "mongolianbeef", "chickenwings"], \
            "Cars": ["astonmartin", "bugatti", "ferrari", "porsche", "lamborghini", "mclaren", "saleen", "koenigsegg", "maserati", "mercedesbenz"]}

class Hangman:
    def __init__(self):
        self.category = 0
        self.secret_word = []
        self.wrong = []
        self.user_guess_correct = []
        self.solved = False
    
    def start_game(self):
        input("Let's play Hangman! Press any key to continue.")
        print("="*50)
        self.category = int(input("To start the game, please choose a category: \n Animals (1) | Candy (2) | Food (3) | Cars (4) "))
        self.pick_word()
        self.start_guesses()

    def pick_word(self):
        word = None
        if self.category == 1:
            print("You chose the Animals category.")
            word = wordDict["Animals"]
        elif self.category == 2:
            print("You chose the Candy category.")
            word = wordDict["Candy"]
        elif self.category == 3:
            print("You chose the Food category.")
            word = wordDict["Food"]
        elif self.category == 4:
            print("You chose the Cars category.")
            word = wordDict["Cars"]
        else:
            print("You entered an invalid category. Please try again!")
        self.secret_word = random.choice(word)
        self.user_guess_correct = list(self.init_user_guess_correct())
    
    def start_guesses(self):
        while len(self.wrong) < 7:
            self.get_guess()
            if self.user_guess_correct.count("_") == 0:
                print(self.user_guess_correct)
                print("You win!")
                return
        print("You're all out of guesses. Better luck next time!")
    
    def get_guess(self):
        print(self.user_guess_correct)
        if len(self.wrong) != 0:
            print("Incorrect guesses:")
            print(self.wrong)
        guess = input("Please guess a letter:")
        self.guess_checker(guess)
        

    def guess_checker(self, guess):
        y = False
        x = 0
        for letter in self.secret_word:
            if letter == guess:
                self.user_guess_correct[x] = guess
                y = True
            x = x + 1    
        if y == False:
            self.wrong.append(guess)
            print("Sorry, that guess wasn't correct. Please try again.")
        else:
            print("Correct! Nice job!!!")

    def init_user_guess_correct(self):
        sw_len = len(self.secret_word)
        blank = "_" * sw_len
        return blank

def main():

    done = False
    while not done:
        start_hangman = Hangman()
        start_hangman.start_game()

        choice = input("Would you like to play again? (Y/N)").lower()
        if choice == 'y':
            continue
        else:
            confirm = input('Are you sure you want to quit? Y/N ? ').lower()
            if confirm == 'y':
                done = True
            elif confirm == 'n':
                continue

main()        

    
```
